
## Основной редактор + морфинг


```python

import os, json, glob
import numpy as np
import cv2
from IPython.display import Video, display
import matplotlib.pyplot as plt
from matplotlib.collections import PathCollection
import ipywidgets as widgets
from dataclasses import dataclass

# SciPy Delaunay (опционально)
try:
    from scipy.spatial import Delaunay as SciPyDelaunay
    SCIPY_AVAILABLE = True
except Exception:
    SCIPY_AVAILABLE = False

# === Config ===
SRC = "img/08.JPG"     # ДО
TGT = "img/09.JPG"     # ПОСЛЕ
COMMON_SIZE = (640, 480)  # (width, height) или None
PTS_JSON = "points_08_to_09.json"
OUT_DIR = "out_orthomorph"
OUT_DIR_FRAMES = os.path.join(OUT_DIR, "frames")
os.makedirs(OUT_DIR_FRAMES, exist_ok=True)
OUT_VIDEO = os.path.join(OUT_DIR, "orthomorph.mp4")
N_STEPS = 50
FPS = 30
QUALITY_JPG = 95
AUTO_BOUNDARY = True
BOUNDARY_MARGIN = 0

# === I/O ===
def imread_rgb(path):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    if img is None:
        raise FileNotFoundError(f"Cannot read image: {path}")
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def resize_to_common(img, target_size):
    if target_size is None: return img
    w, h = target_size
    return cv2.resize(img, (w, h), interpolation=cv2.INTER_AREA)

def prepare_images(src_path, tgt_path, common_size=COMMON_SIZE):
    imgA = imread_rgb(src_path)
    imgB = imread_rgb(tgt_path)
    if common_size is not None:
        imgA = resize_to_common(imgA, common_size)
        imgB = resize_to_common(imgB, common_size)
    else:
        h, w = imgA.shape[:2]
        imgB = cv2.resize(imgB, (w, h), interpolation=cv2.INTER_AREA)
    assert imgA.shape == imgB.shape, "Images must have same size"
    return imgA, imgB

# === Geometry/Morph ===
def apply_affine_transform(src, src_tri, dst_tri, size):
    warp_mat = cv2.getAffineTransform(np.float32(src_tri), np.float32(dst_tri))
    return cv2.warpAffine(src, warp_mat, (size[0], size[1]), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT_101)

def rect_contains(rect, point):
    x, y, w, h = rect
    return (x <= point[0] < x + w) and (y <= point[1] < y + h)

def warp_triangle(img, img_warp, t_src, t_dst):
    r1 = cv2.boundingRect(np.float32([t_src]))
    r2 = cv2.boundingRect(np.float32([t_dst]))
    t1_rect = [((t_src[i][0] - r1[0]), (t_src[i][1] - r1[1])) for i in range(3)]
    t2_rect = [((t_dst[i][0] - r2[0]), (t_dst[i][1] - r2[1])) for i in range(3)]
    mask = np.zeros((r2[3], r2[2], 3), dtype=np.float32)
    cv2.fillConvexPoly(mask, np.int32([t2_rect]), (1.0, 1.0, 1.0))
    img1_rect = img[r1[1]:r1[1]+r1[3], r1[0]:r1[0]+r1[2]]
    size = (r2[2], r2[3])
    warp = apply_affine_transform(img1_rect, t1_rect, t2_rect, size)
    img_warp[r2[1]:r2[1]+r2[3], r2[0]:r2[0]+r2[2]] =         img_warp[r2[1]:r2[1]+r2[3], r2[0]:r2[0]+r2[2]] * (1 - mask) + warp * mask

def add_boundary_points(points, w, h, margin=0):
    m = margin
    pts = list(points)
    pts += [(0+m,0+m),(w-1-m,0+m),(w-1-m,h-1-m),(0+m,h-1-m)]
    pts += [(w//2,0+m),(w//2,h-1-m),(0+m,h//2),(w-1-m,h//2)]
    return pts

def delaunay_tris(points, w, h):
    pts = np.array(points, dtype=np.float32)
    if SCIPY_AVAILABLE:
        return SciPyDelaunay(pts).siplaces.tolist() if hasattr(SciPyDelaunay(pts),'siplaces') else SciPyDelaunay(pts).simplices.tolist()
    rect = (0, 0, w, h)
    subdiv = cv2.Subdiv2D(rect)
    for p in pts: subdiv.insert((float(p[0]), float(p[1])))
    triangle_list = subdiv.getTriangleList()
    triangles = []
    for t in triangle_list:
        x1,y1,x2,y2,x3,y3 = t
        p1=(x1,y1); p2=(x2,y2); p3=(x3,y3)
        if rect_contains((0,0,w,h),p1) and rect_contains((0,0,w,h),p2) and rect_contains((0,0,w,h),p3):
            def nearest_idx(pt):
                d = np.sum((pts-pt)**2, axis=1)
                return int(np.argmin(d))
            tri_idx = sorted(list({nearest_idx(p1),nearest_idx(p2),nearest_idx(p3)}))
            if len(tri_idx)==3: triangles.append(tri_idx)
    uniq, seen = [], set()
    for a,b,c in triangles:
        key = tuple(sorted((a,b,c)))
        if key not in seen:
            seen.add(key); uniq.append([a,b,c])
    return uniq

def morph_images(imgA, imgB, ptsA, ptsB, alpha, triangles=None, add_boundary=True, boundary_margin=0):
    h, w = imgA.shape[:2]
    assert imgB.shape[:2]==(h,w)
    assert len(ptsA)==len(ptsB) and len(ptsA)>=3
    A = [tuple(map(float,p)) for p in ptsA]
    B = [tuple(map(float,p)) for p in ptsB]
    if add_boundary:
        A = add_boundary_points(A, w, h, boundary_margin)
        B = add_boundary_points(B, w, h, boundary_margin)
    P = [((1-alpha)*A[i][0] + alpha*B[i][0], (1-alpha)*A[i][1] + alpha*B[i][1]) for i in range(len(A))]
    if triangles is None:
        triangles = delaunay_tris(P, w, h)
    out = np.zeros_like(imgA, dtype=np.float32)
    for tri in triangles:
        x,y,z = tri
        t  = [P[x], P[y], P[z]]
        tA = [A[x], A[y], A[z]]
        tB = [B[x], B[y], B[z]]
        tmpA = out.copy()
        tmpB = out.copy()
        warp_triangle(imgA, tmpA, tA, t)
        warp_triangle(imgB, tmpB, tB, t)
        out = (1 - alpha) * tmpA + alpha * tmpB
    return np.clip(out,0,255).astype(np.uint8), triangles

def render_sequence(imgA, imgB, ptsA, ptsB, n_steps=50, out_dir="out_orthomorph/frames", fps=30, quality=95):
    os.makedirs(out_dir, exist_ok=True)
    h,w = imgA.shape[:2]
    mid = [((ptsA[i][0]+ptsB[i][0])/2.0, (ptsA[i][1]+ptsB[i][1])/2.0) for i in range(len(ptsA))]
    mid = add_boundary_points(mid, w, h, BOUNDARY_MARGIN) if AUTO_BOUNDARY else mid
    tris = delaunay_tris(mid, w, h)
    paths = []
    for i,k in enumerate(np.linspace(0,1,n_steps)):
        frame,_ = morph_images(imgA,imgB,ptsA,ptsB,alpha=float(k),triangles=tris,add_boundary=AUTO_BOUNDARY,boundary_margin=BOUNDARY_MARGIN)
        p = os.path.join(out_dir, f"orthomorph_{i:04d}.jpg")
        cv2.imwrite(p, cv2.cvtColor(frame, cv2.COLOR_RGB2BGR), [int(cv2.IMWRITE_JPEG_QUALITY), quality])
        paths.append(p)
    return paths

def build_video_from_frames(pattern, out_path="out_orthomorph/orthomorph.mp4", fps=30):
    files = sorted(glob.glob(pattern))
    if not files: raise RuntimeError("No frames found")
    frm0 = cv2.imread(files[0], cv2.IMREAD_COLOR)
    h,w = frm0.shape[:2]
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    vw = cv2.VideoWriter(out_path, fourcc, fps, (w,h))
    if not vw.isOpened(): raise RuntimeError("Cannot open VideoWriter")
    for f in files:
        im = cv2.imread(f, cv2.IMREAD_COLOR)
        if im is None: continue
        if im.shape[:2]!=(h,w): im = cv2.resize(im, (w,h))
        vw.write(im)
    vw.release()
    return out_path
```

```python

# === Editor UI ===

@dataclass
class EditorState:
    imgA: np.ndarray = None
    imgB: np.ndarray = None
    ptsA: list = None
    ptsB: list = None
    add_pair_mode: bool = True
    drag_mode: bool = False
    selected_side: str = None
    selected_idx: int = -1
    tmp_pair: list = None
    scatterA: PathCollection = None
    scatterB: PathCollection = None

state = EditorState(ptsA=[], ptsB=[], tmp_pair=[])

imgA, imgB = prepare_images(SRC, TGT, COMMON_SIZE)
state.imgA = imgA
state.imgB = imgB

def draw_points(ax, pts):
    if not pts: return ax.scatter([], [])
    arr = np.array(pts)
    sc = ax.scatter(arr[:,0], arr[:,1], s=32, marker='o')
    for i,(x,y) in enumerate(pts):
        ax.text(x+3, y+3, str(i), fontsize=8, color='yellow')
    return sc

def refresh_canvas(fig, axA, axB):
    axA.clear(); axB.clear()
    axA.imshow(state.imgA); axA.set_title("Image A (SRC)")
    axB.imshow(state.imgB); axB.set_title("Image B (TGT)")
    state.scatterA = draw_points(axA, state.ptsA)
    state.scatterB = draw_points(axB, state.ptsB)
    axA.set_axis_off(); axB.set_axis_off()
    fig.canvas.draw_idle()

def nearest_point_idx(pts, xy, max_dist=15.0):
    if not pts: return None
    arr = np.array(pts)
    d = np.sqrt(((arr[:,0]-xy[0])**2 + (arr[:,1]-xy[1])**2))
    i = int(np.argmin(d))
    return i if d[i] <= max_dist else None

status_out = widgets.Output()
btn_reset = widgets.Button(description="Reset points", button_style="warning")
btn_undo  = widgets.Button(description="Undo last pair")
btn_save  = widgets.Button(description="💾 Save JSON", button_style="success")
btn_load  = widgets.Button(description="📂 Load JSON")
toggle_add_pair = widgets.ToggleButton(value=True, description="Add Pair Mode", icon="plus")
toggle_drag = widgets.ToggleButton(value=False, description="Drag Mode", icon="arrows")

def on_click(event):
    if event.inaxes is None or event.xdata is None or event.ydata is None:
        return
    x, y = float(event.xdata), float(event.ydata)
    side = 'A' if event.inaxes is axA else ('B' if event.inaxes is axB else None)
    if side is None: return

    if state.drag_mode:
        pts = state.ptsA if side=='A' else state.ptsB
        if not pts: return
        idx = nearest_point_idx(pts, (x,y), max_dist=25.0)
        if idx is not None:
            state.selected_idx = idx
            state.selected_side = side
            with status_out: print(f"Selected {side}[{idx}] — drag with mouse move.")
        return

    if state.add_pair_mode:
        if side == 'A':
            state.tmp_pair = [(x, y)]
            with status_out: print("Point A registered. Now click corresponding point in B.")
        else:  # B
            if state.tmp_pair and len(state.tmp_pair)==1:
                state.ptsA.append(state.tmp_pair[0])
                state.ptsB.append((x, y))
                state.tmp_pair = []
                with status_out: print(f"Pair added. Total pairs: {len(state.ptsA)}")
            else:
                with status_out: print("Click a point in A first (left image).")
    else:
        if side == 'A': state.ptsA.append((x, y))
        else: state.ptsB.append((x, y))

    refresh_canvas(fig, axA, axB)

def on_motion(event):
    if not state.drag_mode: return
    if state.selected_idx==-1 or state.selected_side is None: return
    if event.inaxes is None or event.xdata is None or event.ydata is None: return
    x,y = float(event.xdata), float(event.ydata)
    if state.selected_side=='A' and 0 <= state.selected_idx < len(state.ptsA):
        state.ptsA[state.selected_idx] = (x, y)
    if state.selected_side=='B' and 0 <= state.selected_idx < len(state.ptsB):
        state.ptsB[state.selected_idx] = (x, y)
    refresh_canvas(fig, axA, axB)

def on_release(event):
    if state.drag_mode and state.selected_idx!=-1:
        state.selected_idx = -1
        state.selected_side = None

def on_key(event):
    if event.key in ('d','backspace'):
        if event.inaxes is None or event.xdata is None or event.ydata is None: return
        side = 'A' if event.inaxes is axA else ('B' if event.inaxes is axB else None)
        if side is None: return
        pts = state.ptsA if side=='A' else state.ptsB
        if not pts: return
        idx = nearest_point_idx(pts, (event.xdata, event.ydata), max_dist=20.0)
        if idx is not None:
            pts.pop(idx)
            with status_out: print(f"Deleted {side}[{idx}] — A={len(state.ptsA)}, B={len(state.ptsB)}")
            refresh_canvas(fig, axA, axB)

def on_btn_reset(b):
    state.ptsA.clear(); state.ptsB.clear(); state.tmp_pair=[]
    refresh_canvas(fig, axA, axB)
    with status_out: print("Points cleared.")

def on_btn_undo(b):
    if state.ptsA and state.ptsB:
        state.ptsA.pop(); state.ptsB.pop()
        refresh_canvas(fig, axA, axB)

def save_points_json(path=PTS_JSON):
    with open(path, "w", encoding="utf-8") as f:
        json.dump({"image_a": SRC, "image_b": TGT, "points_a": state.ptsA, "points_b": state.ptsB}, f, ensure_ascii=False, indent=2)
    with status_out: print(f"Saved to {path}")

def load_points_json(path=PTS_JSON):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    state.ptsA = [tuple(map(float,p)) for p in data.get("points_a",[])]
    state.ptsB = [tuple(map(float,p)) for p in data.get("points_b",[])]
    refresh_canvas(fig, axA, axB)
    with status_out: print(f"Loaded: A={len(state.ptsA)}, B={len(state.ptsB)}")

btn_reset.on_click(on_btn_reset)
btn_undo.on_click(on_btn_undo)
btn_save.on_click(lambda b: save_points_json())
btn_load.on_click(lambda b: load_points_json())
toggle_add_pair.observe(lambda ch: setattr(state, "add_pair_mode", ch["new"]), names='value')
toggle_drag.observe(lambda ch: setattr(state, "drag_mode", ch["new"]), names='value')

fig, (axA, axB) = plt.subplots(1,2, figsize=(12,6), dpi=110)
refresh_canvas(fig, axA, axB)

cid_click   = fig.canvas.mpl_connect('button_press_event', on_click)
cid_motion  = fig.canvas.mpl_connect('motion_notify_event', on_motion)
cid_release = fig.canvas.mpl_connect('button_release_event', on_release)
cid_key     = fig.canvas.mpl_connect('key_press_event', on_key)

display(widgets.HBox([btn_reset, btn_undo, btn_save, btn_load, toggle_add_pair, toggle_drag]))
display(status_out)

print("Instruction:")
print("1) В режиме Add Pair: клик в A, затем соответствующий клик в B — добавит пару точек.")
print("2) В режиме Drag: включите Drag Mode, кликните рядом с точкой и перемещайте мышь.")
print("3) Удаление: наведите курсор и нажмите 'd' или Backspace.")
```

```python

# === Run morphing ===
ptsA = state.ptsA.copy()
ptsB = state.ptsB.copy()
print(f"Using pairs: {len(ptsA)}")
if len(ptsA) != len(ptsB) or len(ptsA) < 3:
    raise RuntimeError("Точек недостаточно или количество не совпадает: нужно >=3 и одинаковое число в A и B.")

# очистим старые кадры
for p in glob.glob(os.path.join(OUT_DIR_FRAMES, "orthomorph_*.jpg")):
    try: os.remove(p)
    except: pass

paths = render_sequence(state.imgA, state.imgB, ptsA, ptsB, n_steps=N_STEPS, out_dir=OUT_DIR_FRAMES, fps=FPS, quality=QUALITY_JPG)
video_path = build_video_from_frames(os.path.join(OUT_DIR_FRAMES, "orthomorph_*.jpg"), out_path=OUT_VIDEO, fps=FPS)
print("Video ready:", video_path)
display(Video(OUT_VIDEO, embed=True))
```
